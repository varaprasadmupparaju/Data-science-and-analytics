# Task #4: To Explore Decision Tree Algorithm

# Importing Libraries

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

# Load dataset

```python
iris = load_iris()
x=pd.DataFrame(iris.data[:,:], columns=iris.feature_names[:])
y=pd.DataFrame(iris.target, columns=["Species"])
y.sample(10)#setosa flowers are identified by zero, versicolor by one, and virginica by two.
```

# Applying Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier(max_depth=6) 
decision_tree.fit(x, y)  
```

# visualizing the decision tree using plot_tree

The plot_tree method easily produce figure of the tree.It requires matplotlib to be installed.

```python
import matplotlib.pyplot as plt
from sklearn import tree
fig = plt.figure(figsize=(40,50))
viz = tree.plot_tree(decision_tree, 
                   feature_names=x.columns,  
                   class_names=iris.target_names,
                   filled=True)
```
